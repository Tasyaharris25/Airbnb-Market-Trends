```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Pricing analysis

#analyze price distribution by room type
price_by_room_type = df.groupby("room_type").agg(
    avg_price = ('price','mean')
)

print("Price for each room type :")
print( price_by_room_type)
print("\n")

#analyze price distribution by property type
price_by_property_type = df.groupby("property_type")['price'].mean()
print("Price for each property type :")
print(price_by_property_type)

print("\n")

#analyze price by bedroom
price_by_room_type = df.groupby("bedrooms")['price'].mean()
print("Property price by number of bedroom :",price_by_room_type)
```

```python
#how many listing by bed
bedrooms_count = df.groupby('bedrooms')['id'].count().sort_values(ascending=False)
print(bedrooms_count)
```

```python
#0 bedrooms

zero_bedroom_df =df[df['bedrooms']==0]

print(zero_bedroom_df.groupby('room_type')['id'].count())
print(zero_bedroom_df.groupby('property_type')['id'].count().sort_values(ascending=False))
```

```python
#Popular features

#how many rentals for each room type
room_type_count = df.groupby('room_type')['id'].count().sort_values(ascending=False)
print(room_type_count)

#how many rentals for each property type
property_type_count = df.groupby('property_type')['id'].count().sort_values(ascending=False)
print(property_type_count)
```

```python
#geograhic trends
#scatter plot of latitude and logitude colored by price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='price', palette='viridis')
plt.title("Property Locations by Price")
plt.show()

# Average price by neighborhood
price_by_neighbourhood = df.groupby('neighbourhood')['price'].mean()
print("Average Price by Neighborhood:\n", price_by_neighbourhood)
```

```python
#customer preference

#filter df with null score value rating
filtered_df= df[df['review_scores_rating'].notnull()]

#analyze review scores by room type
review_by_room_type = filtered_df.groupby("room_type")['review_scores_rating'].mean()
print("Review score rating by room type")
print(review_by_room_type)

print("\n")
#count the popularity of room types
room_type_count = df['room_type'].value_counts()
print("Room type popularity")
print(room_type_count)
```

```python
#does cancellation policy affect the rental rate?

#how many rentals for every cancellation policy?
cancellation_policy_rental_count = df.groupby('cancellation_policy')['id'].count().sort_values(ascending=False)
print(cancellation_policy_rental_count)
```

```python
#boxplot to visualize how rental rates price vary across different cancellation_policy categories.
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='cancellation_policy', y='price')
plt.title("Rental Rates by Cancellation Polict")
plt.xticks(rotation=45)
plt.show()
```

```python
#conducting anova test
#An Analysis of Variance (ANOVA) test is a statistical tool used to compare the means of multiple groups to determine if there are any significant differences between them. ANOVA can help determine if differences are due to random chance or if they are meaningful.
from scipy.stats import f_oneway

groups = [group['price'].dropna() for name, group in df.groupby('cancellation_policy')]
anova_result = f_oneway(*groups)
print("ANOVA Test for Cancellation Policy:")
print(f"F-statistic: {anova_result.statistic:.4f}, p-value: {anova_result.pvalue:.4f}")
```

```python
#does property with cleaning fee affect the rental rate?

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='cleaning_fee', y='price')
plt.title("Rental Rates by Cleaning Fee")
plt.xticks([0,1],['No cleaning fee', 'Has cleaning fee'])
plt.show()
```

```python
#how many rentals by cleaning fee
cleaning_fee_count = df.groupby('cleaning_fee')['id'].count().sort_values(ascending=False)
print(cleaning_fee_count)
```

```python
#Independent two-sample t-test: To determine if two groups differ from each other.
from scipy.stats import ttest_ind

# Separate the groups: Has cleaning fee (True) and no cleaning fee (False)
group1 = df[df['cleaning_fee'] == True]['price'].dropna()  # Properties with cleaning fee
group2 = df[df['cleaning_fee'] == False]['price'].dropna()  # Properties without cleaning fee

# Perform T-test
t_test_result = ttest_ind(group1, group2, equal_var=False)  # Use Welch's t-test

print("T-test for Cleaning Fee:")
print(f"T-statistic: {t_test_result.statistic:.4f}, p-value: {t_test_result.pvalue:.4f}")

# Interpretation
if t_test_result.pvalue < 0.05:
    print("Cleaning fee significantly affects the rental rate.")
else:
    print("No significant effect of cleaning fee on the rental rate.")
```

```python
#is the property with instant bookable system has more rental rate
instant_booking_count = df.groupby('instant_bookable')['id'].count().sort_values(ascending=False)
print(instant_booking_count)
```

```python
#visualize the relationship between instant bookable and rental rate
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='instant_bookable', y='price')
plt.title("Rental Rates by Instant Bookable")
```

```python
#perform t-test
from scipy.stats import ttest_ind
group1 = df[df['instant_bookable'] == 't']['price'].dropna()  # Properties with instant booking
group2 = df[df['instant_bookable'] == 'f']['price'].dropna()  # Properties without instant
t_test_result = ttest_ind(group1, group2, equal_var=False) #use Welch's t-test

print("T-test for Instant Bookable:")
print(f"T-statistic: {t_test_result.statistic:.4f}, p-value: {t_test_result.pvalue:.4f}")

#interpretation
if t_test_result.pvalue < 0.05:
    print("Instant booking significantly affects the rental rate")
else:
    print("No significant effect of instant booking on the rental rate")
```
